####3. Use Pandas "Read JSON" functionality to turn the 311 data into a Pandas Dataframe and display the first 20 rows


```python
#YOUR CODE HERE

# Use Pandas "Read JSON" functionality to turn the 311 data into a Pandas Dataframe:
pd_data = pd.read_json(url)

# Display the first 20 rows in the dataframe:
pd_data[:20]
```

####4. How many missing values do we have? 

```python
#YOUR CODE HERE

# Each column has a different number of missing values, which can be observed here:
n_nan = np.sum(pd.isnull(pd_data))
print(n_nan)

# The total number of missing values is:
n_nan = n_nan * 1;
n_nan = np.sum(n_nan[:])
n_nan = np.sum(n_nan)

print("The number of missing values is: " + str(n_nan))
```
