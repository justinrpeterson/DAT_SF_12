####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE

# Grab initial sequal query:
df_q9 = pd.read_sql("SELECT neighborhood, category, COUNT(*) as countType from " + df_name + " GROUP BY neighborhood, category  ORDER BY countType DESC", con)
df_q9 = df_q9.dropna()

# Get the list of neighborhoods
list_names = pd.unique(df_q9['neighborhood'])

# Make a list of most common complain in each hood:
hood_list = pd.DataFrame(columns=df_q9.columns)
for ii in range(len(list_names)):
    tempName = pd.DataFrame(df_q9[df_q9['neighborhood']==str(list_names[ii])])
    hood_list = hood_list.append(tempName.iloc[0])
    
# Make a variable where the most common complain is not 'Street and Sidewalk Cleaning':
final_list = pd.DataFrame(columns=df_q9.columns)
for ii in range(len(list_names)):
    xName = hood_list.iloc[ii]
    if 'Street and Sidewalk Cleaning' != str(xName['category']):
        final_list = final_list.append(hood_list.iloc[ii])
    
# Print the results for the solution    
print("The following neighborhoods do not have 'Street and Sidewalk Cleaning' as their main complaint \n")
print(final_list)
#pd.DataFrame.to_csv(final_list, 'Final_list.csv')    
```
