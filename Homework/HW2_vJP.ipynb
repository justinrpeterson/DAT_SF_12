```python
# Load the data and stuff:
%matplotlib inline
import pandas as pd
import numpy as np
from seaborn import plt

# sklearn packages
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score
```

## Supervised (KNN)

```python
# Read 'wine.csv' in the data folder.
wine_loc = '/Users/jp/Documents/introPy/DAT_SF_12/Data/'
wine_df = pd.io.parsers.read_csv(wine_loc + 'wine.csv')
wine_df.Wine.unique()
```

```python
wine_tgt = pd.Series(wine_df.Wine)
```

```python
wine_data = wine_df.drop('Wine',axis=1)
wine_data.head(5)
```

```python
# Split the data into Train and Test (30%) using random_state of 42 (so that we all get same results)
X_train, X_test, y_train, y_test = train_test_split(wine_data.values, wine_tgt.values, test_size=0.3, random_state=42)
```

### Using KNN, can you predict the Wine class in test

```python
## Using KNN, can you predict the Wine class in test

# Various variables we'll need to set intially.
n_neighbors = range(3, 51, 2)

# Loop through each neighbors value from 1 to 51 and append the scores
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```

### How Accurate is your prediction

```python
# plot the scores:
plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
plt.title('Prediction Accuracy in test set across KNN')
plt.show()
```

```python
# get the max score
max(scores)
k = scores.index(max(scores))
```

## Part 2: Un-Supervised (Clustering)
#### The First Column contains the Wine Category.  Don't use it in the models below.  We are going to treat it as unsupervized learning and compare the results to the Wine column.

```python
# import the necessary sklearn kits:
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import scale
```

####Try `KMeans` where `n_clusters = 3` and compare the clusters to the `Wine` column.

```python
# Try `KMeans` where `n_clusters = 3` and compare the clusters to the `Wine` column.
n_clusters = 3
wine_data_scale = pd.DataFrame(scale(wine_data))
wine_data_scale.describe()

# Test the k-means
kmeans = KMeans(n_clusters, random_state=2)
yHat_kmeans = kmeans.fit(wine_data).labels_
yHat_kmeans = yHat_kmeans + 1

prct = sum((yHat_kmeans == wine_tgt) * 1.00) / wine_tgt.count()
print(prct)
```

```python
# Plot the kmeans data
y = pd.Series(wine_tgt)
yHat = pd.Series(yHat_kmeans)
comp_data = pd.DataFrame(pd.concat([y,yHat],axis=1))
comp_data.columns = ['y', 'yHat']
plt.plot(comp_data)
plt.show()
```

#### Check to se if scaling the data appears to improve the fit:

```python
kmeans = KMeans(n_clusters, random_state=2)
yHat_kmeans = kmeans.fit(wine_data_scale).labels_
yHat_kmeans = yHat_kmeans + 1

prct = sum((yHat_kmeans == wine_tgt) * 1.00) / wine_tgt.count()
print(prct)
```

#### Try `Hierarchical Clustering` and plot dendogram

```python
# no see if we can fit the clusters:
link_type = 'average'
hierarch = AgglomerativeClustering(n_clusters=n_clusters, linkage=link_type)
y_hat = hierarch.fit(wine_data).labels_

# not printed as pretty, but the values are correct
distx = squareform(pdist(wine_data_scale, metric='euclidean'))
```

```python
# plot
R = dendrogram(linkage(distx, method='ward'), color_threshold=80)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram - Scaled Wine Data', fontweight='bold', fontsize=14);
print('Using the dedogram and scaled data I have chosen to use 4 clustering')
```
